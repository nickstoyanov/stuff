```python
# import necessary libraries
#for structuring of data
import pandas as pd
import numpy as np
#for image creation
from PIL import Image
import matplotlib.pyplot as plt
import cv2
from wordcloud import WordCloud, STOPWORDS
```

Нека заредим и разгледаме данните. Данните бяха предварително трансформирани в preprocessing notebook-а.


```python
# Dataset
dataset = pd.read_csv("reduced.csv")
print(dataset.shape)
dataset.info()
dataset.head()
```

```python
genres = dataset.genre.unique()
print(genres)
```

```python
def createImageGenre(genreName):
    count = len(genreName)
    img = np.zeros((640, count * 480, 3), np.uint8)
    font = cv2.FONT_HERSHEY_TRIPLEX
    cv2.putText(img, genreName, (0, 500), font, 18, (255,255,255), 85)
    img = 255 - img
    
    cv2.imwrite(genreName + ".jpg", img)
    return img

for genre in genres:
    plt.imshow(createImageGenre(genre.upper()))
```

```python
def plot_wordcloud(words, img_path):
    img = Image.open(img_path)
    img_mask = np.array(img)
    plt.figure(figsize=(20,16))
    wc = WordCloud(background_color="black", max_words=300, mask=img_mask,
                   stopwords=STOPWORDS, min_font_size=10, max_font_size= 80)
    wc.generate(" ".join(words))
    plt.imshow(wc.recolor(colormap= 'Pastel1_r' , random_state=17), alpha=20)
    plt.axis('off')
    

words = dataset[dataset.genre=="Pop"].lyrics.values
file_path = "Pop".upper() + ".jpg"
plot_wordcloud(words, file_path)
```
